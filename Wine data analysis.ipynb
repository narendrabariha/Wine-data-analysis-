# Numerical Analysis of Columns

## Histogram plot

```python
plt.figure(figsize=(6,3)) # (8,6) --("width","height")
# Histogram plot
sns.histplot(df["pH"],bins=30) # sns.histplot(df["pH"],kde=True)

plt.title("Histogram plot for pH",y=1.02)
plt.xlabel("pH")
plt.ylabel("Frequency")
plt.show()
```

# Distribution plot

```python
plt.figure(figsize=(6,3))
# Distribution plot
sns.displot(df["pH"],kind="kde") # or sns.kdeplot(df["pH"])
# kde stands for kernel density estimation

plt.title("Density plot for pH",y=1.02)
plt.xlabel("pH")
plt.ylabel("Probability")
plt.show()
```

###  Are there any extreme values of pH, if so what are they?

```python
df["pH"].quantile(0.10),df["pH"].quantile(0.25),df["pH"].quantile(0.50),df["pH"].quantile(0.75),df["pH"].quantile(0.90)
```

# Box plot

```python
plt.figure(figsize=(6,3))
# Box plot
sns.boxplot(df["pH"],orient="h") # orient="h" for horizontal or orient="v" for vertical

plt.title("Box plot for pH",y=1.02)
plt.xlabel("pH")
# plt.ylabel("")
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(8,5))
# Violin Plot
sns.violinplot(df["pH"],orient="h",color="tomato") # refer color guide at bottom
sns.swarmplot( df["pH"],size=0.5,orient="h")

plt.title("Violin plot for pH",y=1.02)
plt.xlabel("pH")
# plt.ylabel("")
plt.show()
```

 **pH** Column has a symmetrical distribution with skew 0.38 which is between -0.5 and 0.5 .

- Density curve for pH is centred around mean 2.6 with standard deviation of 2.8
- Range of pH value is from minimum 2.72 and maximum 4.01
- Most of the people are failling under pH group of 3.1 - 3.3
- 10% of people are of pH 3.02 and below.
- 25% of people are of pH 3.11 and below.
- 50% of people are of pH 3.21 and below.
- 75% of people are of pH 3.32 and below.
- 90% of people are of pH 3.42 and below.
- pH column has few extreme values toward higher pH group 3.7 and above. 

pH column could be a good predictor for wine and its relation with other columns should be explored further.
